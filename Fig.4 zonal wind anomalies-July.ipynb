```python
import iris 
import matplotlib.pyplot as plt
import iris.quickplot as qplt
import iris.plot as iplt
import numpy as np

%matplotlib inline
```

```python
data_dir = '/shared/netscratch/lrm49/UKCA_VN112_QBO_SIMS/'
data_dir1 = '/shared//netscratch/lrm49/UKCA_VN112_QBO_SIMS/PERTLIM_CONTROLS/'
```

```python

JLW15 = iris.load(data_dir + 'JULY_W/u-bn945_all_vars_processed.nc', 'x_wind')
JLW60 = iris.load(data_dir + 'JULY_W/u-bn515_all_vars_processed.nc', 'x_wind')
JLE15 = iris.load( data_dir + 'JULY_E/u-bo057_all_vars_processed.nc', 'x_wind')
JLE60 = iris.load(data_dir + 'JULY_E/u-bo005_all_vars_processed.nc', 'x_wind')
```

```python

JUL_E= iris.load(data_dir+'JULY_E/u-bl775_all_vars_processed.nc', 'x_wind')
JLE1 = iris.load(data_dir1+'u-br015_all_vars_processed.nc', 'x_wind')
JLE2 = iris.load(data_dir1+'u-br016_all_vars_processed.nc', 'x_wind')
JUL_W = iris.load(data_dir+'JULY_W/u-bl775_all_vars_processed.nc', 'x_wind')
JLW1 = iris.load(data_dir1+'u-br011_all_vars_processed.nc', 'x_wind')
JLW2 = iris.load(data_dir1+'u-br012_all_vars_processed.nc', 'x_wind')
```

```python
list_x = [JLW15, JLW60, JLE15, JLE60]
```

```python
control_list = [ JUL_W, JLW1,JLW2, JUL_E, JLE1, JLE2]
```

```python
latlist = []
for eruption in list(list_x):
    lat_constraint = eruption.extract(iris.Constraint(latitude = lambda cell: -2 <= cell <=2))
    latlist.append(lat_constraint)
```

```python
latlist_cont = []
for eruption in control_list:
    lat_constraint = eruption.extract(iris.Constraint(latitude = lambda cell: -2 <= cell <=2))
    latlist_cont.append(lat_constraint)
```

```python
import iris.analysis.cartography
```

```python

latlist[0][0].coord('latitude').guess_bounds()
latlist[0][0].coord('longitude').guess_bounds()
```

```python
grid_areas = iris.analysis.cartography.area_weights(latlist[0][0])
```

```python
areaav_list = []
for eruption in latlist:
    area_average = eruption[0].collapsed(['latitude', 'longitude'], iris.analysis.MEAN, weights = grid_areas)
    areaav_list.append(area_average)
```

```python
areaav_list_cont = []
for eruptions in latlist_cont:
    for eruption in eruptions:
        area_average = eruption.collapsed(['latitude', 'longitude'], iris.analysis.MEAN)
        areaav_list_cont.append(area_average)
```

```python
controls = [eruption for eruption in areaav_list_cont]
```

```python
control_list = [controls[0:3], controls[3:6]]
```

```python
control_mean = [sum(ensembles)/3 for ensembles in control_list]
```

```python
eruptions = [eruption for eruption in areaav_list]
```

```python
datasetW = [control_list[0][0].data - control_list[0][1].data, control_list[0][1].data - control_list[0][2].data, control_list[0][0].data - control_list[0][2].data]
```

```python
datasetE = [control_list[1][0].data - control_list[1][1].data, control_list[1][1].data - control_list[1][2].data, control_list[1][0].data - control_list[1][2].data]
```

```python
control_list[0]
```

```python
w_60 = [eruptions[1].data - control.data for control in control_list[0]]
w_15 = [eruptions[0].data - control.data for control in control_list[0]]
e_60 = [eruptions[3].data - control.data for control in control_list[1]]
e_15 = [eruptions[2].data - control.data for control in control_list[1]]
```

```python
import scipy
```

```python
stats, pW = scipy.stats.ttest_ind(np.array(datasetW), np.array(w_60), axis=0)
bool_matrix_W60 = np.where( pW > 0.05, 1, 0)
```

```python
stats, pE = scipy.stats.ttest_ind(np.array(datasetE), np.array(e_60), axis=0)
bool_matrix_E60 = np.where( pE > 0.05, 1, 0)
```

```python
stats, pw = scipy.stats.ttest_ind(np.array(datasetW), np.array(w_15), axis=0)
bool_matrix_W15 = np.where( pw > 0.05, 1, 0)
```

```python
stats, pe = scipy.stats.ttest_ind(np.array(datasetE), np.array(e_15), axis=0)
bool_matrix_E15 = np.where( pe > 0.05, 1, 0)
```

```python
anomaly_list = []
for i in range(0,2):
    
    anomaly60 = eruptions[(i*2)+1] - control_mean[i]
    anomaly15 = eruptions[i*2] - control_mean[i]
    anomaly_list.append(anomaly15)
    anomaly_list.append(anomaly60)
    
```

```python
wind_anomaly = range(-45,50,10)
titles = [' w-QBO', ' e-QBO']
letter_labels = ('(a)', '(b)', '(c)', '(d)', '(e)', '(f)', 'g)', 'h)', 'i)','j)' ,'k)' ,'l)' )
zero = 0
hatch = np.linspace(0,1,3)
```

```python
matrices = [bool_matrix_W15,  bool_matrix_E15, bool_matrix_W60, bool_matrix_E60]
```

```python
plot_list = [anomaly_list[0], anomaly_list[2], anomaly_list[1], anomaly_list[3]]
line_list = [areaav_list[0], areaav_list[2], areaav_list[1], areaav_list[3]]
```

```python
cm = 1/2.54
fig, axs = plt.subplots(2,2, sharey = True, sharex = False, figsize = (21*cm,28*cm))
plt.rcParams.update({'hatch.color': 'grey'})

ax = axs.ravel()

for i, eruption in enumerate(plot_list):
        x = range(0, 36)
        y = anomaly_list[0].coord('pressure').points
        z = np.transpose(eruption.data)
        z2 = np.transpose(line_list[i].data)
        ax[i].contour(x,y,z2,zero, linewidths = 4, colors = 'darkslategrey')
        cf = ax[i].contourf(x,y,z,wind_anomaly, extend = 'both', cmap = plt.cm.bwr, alpha = 0.65)
        ax[i].contourf(x,y,np.transpose(matrices[i]),hatch, colors = 'none', hatches = [None, '..'])
        ax[i].set_yscale('log')
        
        ax[i].plot(0,65, 'r^',  clip_on=False, markersize = 20, markeredgecolor = 'k', markeredgewidth = 1, zorder = 100)
        
    
for i in range(0,4):

    ax[i].xaxis.set_tick_params(width = 1.5, labelsize = 13, direction = 'out', length = 3)
    ax[i].yaxis.set_tick_params(which = 'both', width = 1.5, labelsize = 13, direction = 'out', length = 3)
    ax[i].set_ylim(1,100)
    ax[i].invert_yaxis()

ax[0].set_title(titles[0],loc = 'center', fontsize = 20, y = 1.2, fontweight = 'bold') 

ax[1].set_title(titles[1],loc = 'center', fontsize = 20, y = 1.2, fontweight = 'bold') 


ax[0].set_ylabel('Pressure / hPa', fontsize = 15, labelpad = 10)
ax[2].set_ylabel('Pressure / hPa', fontsize = 15, labelpad = 10)
ax[0].set_title(letter_labels[0] + '     15 Tg of SO$_2$', size=18,  y = 1.03, loc = 'left')
ax[1].set_title(letter_labels[1] + '     15 Tg of SO$_2$', size=18,  y = 1.03, loc = 'left')
ax[2].set_title(letter_labels[2] + '     60 Tg of SO$_2$', size=18,  y = 1.03, loc = 'left')
ax[3].set_title(letter_labels[3] + '     60 Tg of SO$_2$', size=18,  y = 1.03, loc = 'left')
    
ax[2].set_xlabel('Months since eruption', fontsize = 15, labelpad = 15)    
ax[3].set_xlabel('Months since eruption', fontsize = 15, labelpad = 15)


cbar_ax = fig.add_axes([0.15, 0.1, 0.8, 0.02])
colourbar = plt.colorbar(cf, cax = cbar_ax , orientation = 'horizontal')
colourbar.set_label('m s$^{-1}$', fontsize = 14)
colourbar.ax.tick_params(labelsize=12)


plt.tight_layout() 
plt.subplots_adjust(bottom = 0.22, wspace = 0.1, hspace = 0.35) 
plt.savefig('Fig04.png', dpi = 500)
```
