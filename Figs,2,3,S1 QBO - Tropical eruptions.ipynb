# Constrain data to get equatorial x wind

### Collapse over latitude

```python
latlist = []
for eruption in list_x:
    lat_constraint = eruption.extract(iris.Constraint(latitude = lambda cell: -2 <= cell <=2))
    latlist.append(lat_constraint)
```

```python
latlist_cont = []
for eruption in control_list:
    lat_constraint = eruption.extract(iris.Constraint(latitude = lambda cell: -2 <= cell <=2))
    latlist_cont.append(lat_constraint)
```

### Average over area

```python
import iris.analysis.cartography
```

```python
areaav_list = []
for eruption in latlist:
    eruption[0].coord('latitude').guess_bounds()
    eruption[0].coord('longitude').guess_bounds()
    grid_areas = iris.analysis.cartography.area_weights(eruption[0])
    area_average = eruption[0].collapsed(['latitude', 'longitude'], iris.analysis.MEAN, weights = grid_areas)
    areaav_list.append(area_average)
```

```python
areaav_list_cont = []
for eruption in latlist_cont:
    eruption[0].coord('latitude').guess_bounds()
    eruption[0].coord('longitude').guess_bounds()
    grid_areas = iris.analysis.cartography.area_weights(eruption[0])
    area_average = eruption[0].collapsed(['latitude', 'longitude'], iris.analysis.MEAN, weights = grid_areas)
    areaav_list_cont.append(area_average)
```

### Make control ensemble mean

```python
control_sum = []
stds = []
for i in range(0,12,3):
    summed = areaav_list_cont[i][0:35] + areaav_list_cont[i+1][0:35] + areaav_list_cont[i+2][0:35]
    std = np.std(np.array([areaav_list_cont[i][0:35].data, areaav_list_cont[i+1][0:35].data, areaav_list_cont[i+2][0:35].data]), axis = 0)
    control_mean = summed / 3
    control_sum.append(control_mean)
    stds.append(std)
```

### Design plot

```python
x = np.linspace(0, 35, 35)
names = [ 'Control ensemble mean', 'Control ensemble mean', 'Eruption releasing 15 Tg of SO$_2$', 'Eruption releasing 15 Tg of SO$_2$', 'Eruption releasing 60 Tg of SO$_2$', 'Eruption releasing 60 Tg of SO$_2$',  'Control ensemble mean', '15 Tg', '60 Tg', 'Control ensemble mean', '15 Tg', '60 Tg', ]
eruptions = ['Zonal wind (2$^\circ$N2$^\circ$S): w-QBO', 'Zonal wind (2$^\circ$N2$^\circ$S): e-QBO']
letter_labels = ('(a)', '(b)', '(c)', '(d)', '(e)', '(f)', '(g)', '(h)', '(i)','(j)' ,'(k)' ,'(l)' )
contour_levels = range(-40, 41, 10)
```

## wind shears

```python
## log-pressure calculation
pressures = areaav_list[0].coord('pressure').points
press_array = np.array(pressures)
z_p = 7e3*np.log(1000 / press_array)
```

```python
gradient_x = [np.gradient(simulation.data, z_p, axis = 1) for simulation in areaav_list]
```

```python
gradient_x_c = [np.gradient(simulation.data, z_p, axis = 1) for simulation in control_sum]
```
