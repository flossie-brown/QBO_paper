```python
import iris 
import matplotlib.pyplot as plt
import iris.quickplot as qplt
import iris.plot as iplt
import numpy as np


%matplotlib inline
```

```python
data_dir = '/shared/netscratch/lrm49/UKCA_VN112_QBO_SIMS/'
data_dir1 = '/shared//netscratch/lrm49/UKCA_VN112_QBO_SIMS/PERTLIM_CONTROLS/'
```

```python

JAN_W = iris.load(data_dir+'JAN_W/u-bl775_all_vars_processed.nc')
JNW1 = iris.load(data_dir1+'u-br303_all_vars_processed.nc')
JNW2 = iris.load(data_dir1+'u-br304_all_vars_processed.nc')
a = iris.load(data_dir+'JAN_W/u-bo186_all_vars_processed.nc')


JUL_E= iris.load(data_dir+'JULY_E/u-bl775_all_vars_processed.nc')
JLE1 = iris.load(data_dir1+'u-br015_all_vars_processed.nc')
JLE2 = iris.load(data_dir1+'u-br016_all_vars_processed.nc')
j = iris.load(data_dir+'JULY_E/u-bo005_all_vars_processed.nc')


JUL_W = iris.load(data_dir+'JULY_W/u-bl775_all_vars_processed.nc')
JLW1 = iris.load(data_dir1+'u-br011_all_vars_processed.nc')
JLW2 = iris.load(data_dir1+'u-br012_all_vars_processed.nc')
d = iris.load(data_dir+'JULY_W/u-bn515_all_vars_processed.nc')


JAN_E = iris.load(data_dir+'JAN_E/u-bl775_all_vars_processed.nc')
JNE1 = iris.load(data_dir1+'u-br301_all_vars_processed.nc')
JNE2 = iris.load(data_dir1+'u-br302_all_vars_processed.nc')
g = iris.load(data_dir+'JAN_E/u-bo332_all_vars_processed.nc')
```

```python
master_list = [a,d,g,j]
```

```python
control_list = [JAN_W, JNW1, JNW2, JUL_W,JLW1, JLW2,  JAN_E, JNE1, JNE2,  JUL_E, JLE1, JLE2]
```

### Calculate SAOD averaged over longitude

```python
list_t = []
for cube in master_list:
    x = cube[1]
    list_t.append(x[0:35])
```

```python
list_tc = []
for i, cube in enumerate(control_list):
    if i%3 ==0:
        x = cube[1]
    else:
        x = cube[1]
    list_tc.append(x[0:35])
```

## Find control ensemble averages

```python
tc_avg = [sum(list_tc[i:i+3])/3 for i in range(0,12,3)]
```

### extract tropics, NH, SH

```python
tropics = [cube.extract(iris.Constraint(latitude=lambda cell: -30 <= cell <= 30)) for cube in list_t]
tropics_c = [cube.extract(iris.Constraint(latitude=lambda cell: -30 <= cell <= 30)) for cube in tc_avg]
NH = [cube.extract(iris.Constraint(latitude=lambda cell: 30 <= cell <= 90)) for cube in list_t]
NH_c = [cube.extract(iris.Constraint(latitude=lambda cell: 30 <= cell <= 90)) for cube in tc_avg]
SH = [cube.extract(iris.Constraint(latitude=lambda cell: -90 <= cell <= -30)) for cube in list_t]
SH_c = [cube.extract(iris.Constraint(latitude=lambda cell: -90 <= cell <= -30)) for cube in tc_avg]
```

```python
import iris.analysis.cartography
tropics[0].coord('latitude').guess_bounds()
tropics[0].coord('longitude').guess_bounds()
grid_areas = iris.analysis.cartography.area_weights(tropics[0])

tropics_mean = [cube.collapsed(['longitude', 'latitude'], iris.analysis.MEAN, weights=grid_areas) for cube in tropics]
tropics_mean_c = [cube.collapsed(['longitude', 'latitude'], iris.analysis.MEAN, weights=grid_areas) for cube in tropics_c]

NH[0].coord('latitude').guess_bounds()
NH[0].coord('longitude').guess_bounds()
grid_areas = iris.analysis.cartography.area_weights(NH[0])

NH_mean = [cube.collapsed(['longitude', 'latitude'], iris.analysis.MEAN, weights=grid_areas) for cube in NH]
NH_mean_c = [cube.collapsed(['longitude', 'latitude'], iris.analysis.MEAN, weights=grid_areas) for cube in NH_c]

SH[0].coord('latitude').guess_bounds()
SH[0].coord('longitude').guess_bounds()
grid_areas = iris.analysis.cartography.area_weights(SH[0])

SH_mean = [cube.collapsed(['longitude', 'latitude'], iris.analysis.MEAN, weights=grid_areas) for cube in SH]
SH_mean_c = [cube.collapsed(['longitude', 'latitude'], iris.analysis.MEAN, weights=grid_areas) for cube in SH_c]
```

```python
legend=['Jan w-QBO', 'July w-QBO', 'Jan e-QBO', 'July e-QBO']
color=['firebrick', 'tomato', 'darkblue', 'dodgerblue', 'gold', 'salmon']
dash=['--', '-', '--', '-', 'gold', 'salmon']
```

```python
plt.figure(figsize=(14,7))
plt.rcParams.update({'font.size': 20})
plt.subplot(121)
for i in range(0,4):
    plt.plot(tropics_mean[i].data - tropics_mean_c[i].data, linewidth=4, color=color[i], linestyle=dash[i], label=legend[i])
    plt.xlabel('months since start of simulation')
    plt.ylabel('sAOD at 550 nm')
    plt.title('Tropical sAOD')
    plt.title('(a)', loc='left')
    
plt.legend(fontsize=20)

plt.rcParams.update({'font.size': 13})
plt.subplot(222)
for i in range(0,4):
    plt.plot(SH_mean[i].data - SH_mean_c[i].data, linewidth=4, color=color[i], linestyle=dash[i], label=legend[i])
    plt.xlabel('months since start of simulation')
    plt.ylabel('sAOD at 550 nm')
    plt.title('SH sAOD')
    plt.title('(b)', loc='left')
    
plt.subplot(224)
for i in range(0,4):
    plt.plot(NH_mean[i].data - NH_mean_c[i].data, linewidth=4, color=color[i], linestyle=dash[i], label=legend[i])
    plt.xlabel('months since start of simulation')
    plt.ylabel('sAOD at 550 nm')
    plt.title('NH sAOD')
    plt.title('(c)', loc='left')
    
plt.tight_layout()
plt.savefig('FigSsAOD.png', dpi = 500)
```

```python
lon_avg = [cube.collapsed('longitude', iris.analysis.MEAN) for cube in list_t]
lon_avg_c = [cube.collapsed('longitude', iris.analysis.MEAN) for cube in tc_avg]
```

```python
plt.figure(figsize=(8,5))
for i in range(0,4):
    plt.plot(lon_avg[i].coord('latitude').points, lon_avg[i][5].data - lon_avg_c[i][5].data,
             linewidth=4, color=color[i], linestyle=dash[i], label=legend[i])
    
plt.xlabel('Latitude')
plt.ylabel('sAOD at 550 nm')
plt.legend()

plt.title('(a)', loc='left', pad=100)
plt.title('60 Tg SO$_2$', loc='center')
```
